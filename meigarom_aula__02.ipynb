```
# 5. Criar um mapa indicando onde as casas estão localizadas geograficamente
# Separar as variáveis que serão base do gráfico.

df_mapa = df[['id','lat','long','price']]

df_mapa

mapa = px.scatter_mapbox (df_mapa, lat='lat', lon = 'long',
                          hover_name='id',
                          hover_data=['price'],
                          color_discrete_sequence=['red'],
                          zoom=3,
                          height=300)

mapa.update_layout( mapbox_style='open-street-map')
mapa.update_layout(height=600, margin= {'r':0, 't':0, 'l':0, 'b':0} )
mapa.show()


```

```
# 6. Crie uma nova coluna chamada: "house_age"
# - Se o valor da coluna "date" for maior que 2014-01-01 => 'new_house
# - Se o valor da coluna "date" for menor que 2014-01-01 => 'old_house

# Aqui poderia criar uma coluna na posição de preferência: 
# df.insert(3,'house_age','age')

df['house_age'] = 'age'

df.loc[df['date'] >= '2014-01-01', 'house_age'] = 'new_house'

df.loc[df['date'] <= '2014-01-01', 'house_age'] = 'old_house'

df['house_age']
```

```
# Verificação dos valores para coluna criada 'house_age'!

df['house_age'].value_counts()
```

```
# 7. Crie uma nova coluna chamada: "dormitory_type"
# - Se o valor da coluna "bedrooms" for igual à 1 => 'studio'
# - Se o valor da coluna "bedrooms" for igual à 2 => 'apartament'
# - Se o valor da coluna "bedrooms" for maior que 2 => 'house'.

df['dormitory_type'] = None

df.loc[df['bedrooms'] == 1, 'dormitory_type'] = 'studio'

df.loc[df['bedrooms'] == 2, 'dormitory_type'] = 'apartament' 

df.loc[df['bedrooms'] > 2, 'dormitory_type'] = 'house'

df['dormitory_type']
```

```
# Verificação dos valores para coluna criada 'dormitory_type'!

df['dormitory_type'].value_counts()
```

```
# 8. Crie uma nova coluna chamada: "condition_type"
#  - Se o valor da coluna "condition" for menor ou igual à 2 => 'bad'
#  - Se o valor da coluna "condition" for igual à 3 ou 4 => 'regular'
#  - Se o valor da coluna "condition" for igual à 5 => 'good'


df['condition_type'] = None

df.loc[df['condition'] <= 2, 'condition_type'] = 'bad'

df.loc[(df['condition'] == 3) | (df['condition'] == 4), 'condition_type'] = 'regular' 

df.loc[df['condition'] == 5, 'condition_type'] = 'good'
```

```
# Verificação dos valores para coluna criada 'condition_type'!

df['condition_type'].value_counts()
```

```
# 9. Modifique o TIPO da coluna "condition" para STRING

df['condition'] = df['condition'].astype(str)

# Verificação da transformação:

df.dtypes
```

```
# 10. Delete as colunas "sqft_living15" e "sqft_lot15"


# Remoção da coluna 'sqft_living':
df.drop('sqft_living', axis=1, inplace=True)

# Remoção da coluna 'sqft_lot15':
df.drop('sqft_lot15', axis=1, inplace=True)

# Remoção das duas colunas na mesma linha de comandos:
df.drop(['sqft_living', 'sqft_lot15'], axis=1, inplace=True)

# Verificação das colunas removidas:
df.head()
```

```
#11. Modifique o TIPO da coluna "yr_built" para DATE

df['yr_built'] = pd.to_datetime(df['yr_built'], format='%Y')

df.dtypes
# yr_built          datetime64[ns]
```

```
# 12. Modifique o TIPO da coluna "yr_renovated" para DATE

df['yr_renovated'] = pd.to_datetime(df['yr_renovated'], format='%Y', errors='coerce' )

 # parâmetro errors='coerce' faz com que os valores que não forem feita à conversão se torne NaT "(Not a Time)"".

df.dtypes
# yr_renovated      datetime64[ns]
```

```
# 23. Selecione as colunas: "id = 0", "date = 1", "price = 2", "floors" = 7, "zipcode = 16" pelo método:

# - Direto pelo nome das colunas
# - Pelos índices
# - Pelos índices das linhas e o nome das colunas
# - Índices booleanos
```

```
column = ["id", "date", "price", "floors", "zipcode"]
columnsnum = []
for i, col in enumerate(df.columns.to_list()):
    if col in column:
        columnsnum.append(i)

print(df.iloc[0:20, columnsnum])
print("-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
print(df.loc[0:20, column])
print("-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=-=")
columns = []
for i, col in enumerate(df.columns.to_list()):
    if col in column:
        columns.append(True)
    else:
        columns.append(False)
    
print(df.loc[0:20, columns])
```

```
# 24. Salve um arquivo .csv com somente as colunas do item 10 e 17

columns = []
for i, col in enumerate(df.columns.to_list()):
    if 10 <= i <=17:
        columns.append(True)
    else:
        columns.append(False)
    
new_df = df.loc[:, columns]

new_df.to_csv('/content/drive/MyDrive/new_dataframe.csv', index=False)
```
